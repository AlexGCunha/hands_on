### 1) What is the fundamental ideal behind support vector machines?

In classification taks, it is to create the largest separation area (in terms of their inputs space) between two or more labels, while keeping the least possible ammount of observations inside this area. In regressions it is almost the opposite, creating the smallest possible area that englobes the maximum possible number of instances.

### 2) What is a support vector?

In the previous answer I talked about creating the largest separation area. Support vectors are the instances that define the boundaries of this area (i.e., are instances defining the borders of the area), hence giving support to the area.

### 3) Why is it important to scale the inputs when using SVMs?

If inputs are not scaled, the ones with a higher range of values will almost define the support vectors, since most of the distance between instances will be defined by these inputs.

### 4) Can an SVM classifier output a confidence score when it classifies an instance? What about a probability?

The SVM classifier itself wont generate probabilities, since it will not give a score to instances and only classifies based on whether the input is located in which side of the border (and street/separation area). A confidence may be given if one interpret the distance of a point to the border of the decision boundary as confidence.

### 5) How can you choose between LinearSVC, SVC and SGDClassifier ? 

Linear SVC tends to train faster, but assumes linearity of the decision boundary on inputs. SVC on the other hand allows for non-linearities but, without the kernel-trick, would run slower, but choosing different kernels allows for improved results without so much of a compromise on train time. SGDClassifier dont allow the kernel-trick, but could be used in huge datasets, that do not fit in memory, since it trains with one instance at a time.

### 6) Say you've trained an SVM classifier with and RBF kernel, but it seems to underfit the training set. Should you increase or decrease $\gamma$ (gamma)? What about C?

Your should increase $\gamma$. It determines the area of influence of each instance. Hence, higher gammas will make the decision boundary more wiggly (less linear) and more prone to overfit, but if your problem is underfiting, increasing gamma should be fine. Similarlly, you should increase $C$, since it dictates the the leniency of instances inside the sepparation area. With a higher C, you have a larger separation area (street).
